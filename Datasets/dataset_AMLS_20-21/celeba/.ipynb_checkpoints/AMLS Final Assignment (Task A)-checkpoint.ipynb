```python
def count_n_images(path,extension):
    list_dir = []
    list_dir = os.listdir(path)
    count = 0
    for file in list_dir:
        if file.endswith(extension):
            count += 1
    return count
```

```python
path=directory + "/img/"
extension=".jpg"
```

```python
count_n_images(path,extension)
```
